```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

##### Example:

Consider the signal:

$$
y = 3\sin(2\pi \cdot 10t) +2\cos(2\pi \cdot 20t + \pi/4), \quad t\in [0, 0.1]
$$

Sample it at the rate of $f_{S} = 80 \textrm{ Hz}$.

$N = f_{S}L = 8$ samples.

```python
fs = 80
L = 0.1
N = int(L*fs)
```

```python
ts = np.linspace(0, L, N, endpoint=False)

sin_comp = lambda ts: 3 * np.sin(2 * np.pi * 10 * ts)
cos_comp = lambda ts: 2 * np.cos(2 * np.pi * 20 * ts + np.pi/4)

ys = sin_comp(ts) + cos_comp(ts)
```

```python
ts1 = np.linspace(0, L, N*100)
ys1 = sin_comp(ts1) + cos_comp(ts1)
```

```python
fig, ax = plt.subplots(figsize=(12,4))

ax.scatter(ts, ys, label='Samples')
ax.plot(ts1, ys1, label='Signal')

ax.legend()
plt.show()
```
