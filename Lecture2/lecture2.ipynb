A digital audio file is simply an array of numbers.

Sample from a sinusoid with frequency $261.6$ Hz (middle C) for 2 seconds at the sampling rate of $44100$ Hz.

Then play the audio.

```python
fs = 44100
secs = 2
N = fs * secs
ts = np.linspace(0, secs, N)
ys = np.sin(2 * np.pi * 261.6 * ts)
```

```python
Audio(ys, rate=fs)
```

Sample from a sinusoid with frequencies $174.6$ Hz, $220$ Hz and $261.6$ Hz (F3-A3-C4, F major chord) for 2 seconds at the sampling rate of 44100 Hz.

Play the audio.

```python
f3 = np.sin(2 * np.pi * 174.6 * ts)
a3 = np.sin(2 * np.pi * 220.0 * ts)
c4 = np.sin(2 * np.pi * 261.6 * ts)

f_major = f3 + a3 + c4
```
